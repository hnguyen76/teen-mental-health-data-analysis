```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Teen_Mental_Health_cleaned.csv')
df.head()
```

```python
ax = df['depression_label'].value_counts().plot(kind='bar')
ax.bar_label(ax.containers[0], fmt="%.2f", padding=3)

plt.title('Depression Label Distribution')
plt.xlabel('Depression Label')
plt.ylabel('Count')

plt.show()
```

```python
avg_social = (
    df.groupby('depression_label')['daily_social_media_hours']
    .mean()
).round(2)
avg_social

ax = avg_social.plot(kind='bar')
ax.bar_label(ax.containers[0], fmt="%.2f", padding=3)

plt.title('Average Social Media Hours by Depression')
plt.xlabel('Depression Label')
plt.ylabel('Average Social Media Hours')

plt.plot()
```

```python
df["sleep_hours"].plot(kind='hist', bins=10, edgecolor='black')
plt.title('Sleep Hours Distribution')
plt.xlabel('Sleep Hour')
plt.ylabel('Number of Students')
           
plt.show()
```

```python
plt.scatter(df['stress_level'], df['anxiety_level'], alpha=0.3)

plt.title("Stress vs Anxiety")
plt.xlabel("Stress Level")
plt.ylabel("Anxiety Level")

plt.show()
```

```python
import seaborn as sns

table = pd.crosstab(df["stress_level"], df["anxiety_level"])

plt.figure(figsize=(8, 6))
sns.heatmap(table, cmap="Blues", annot=True, fmt="d")

plt.title("Stress vs Anxiety Count")
plt.xlabel("Anxiety Level")
plt.ylabel("Stress Level")
plt.show()
```

```python
avg_sleep = (
    df.groupby(['depression_label'])['sleep_hours']
    .mean()
)

ax = avg_sleep.plot(kind='bar', edgecolor='black')
ax.bar_label(ax.containers[0], fmt="%.2f", padding=3)

plt.title('Average Sleep Hours by Depression')
plt.xlabel('Depression label')
plt.ylabel('Average Sleep Hours')

plt.show()
```
